```python
file_r = "./data_r.txt"
file_t = "./data_t.txt"
```

```python
df_r = pd.read_csv(file_r, sep=" ", names=["X", "Y", "Z"])
df_t = pd.read_csv(file_t, sep=" ", names=["X", "Y", "Z"])
```

```python
def filter_data(df):
    x = []
    y = []
    z = []
    print(len(df.X))
    for i in range(0 , len(df.X)):
        if(i%4==0):
            x.append(df.X[i])
            y.append(df.Y[i])
            z.append(df.Y[i])
    return np.array(x), np.array(y), np.array(z)
```

```python
x_r,y_r,z_r = filter_data(df_r)
x_t,y_t,z_t = filter_data(df_t)
```
